```python
veiculo = {"veiculos": [], "modelo": [], "placa": []}

def liscad(veiculo):

    while True:
        v = input('Qual é seu veículo? (carro/moto/caminhão): ').strip().lower()

        if v == "carro" or v == "moto" or v == "caminhao":
            break
        else:
            print(f"Tente novamente. Isso não é uma das opções: {v}")

    m = input(f'Qual é o modelo do seu {v}? ').strip().lower()
    p = input(f'Qual é a placa do seu {v}? ').strip().lower()

    veiculo["veiculos"].append(v)
    veiculo["modelo"].append(m)
    veiculo["placa"].append(p)

    print(f" Seu veículo foi cadastrado com sucesso!")

def remover(veiculo):
    if len(veiculo["veiculos"]) == 0:
        print(" Nenhum veículo cadastrado ainda!")
        return

    print("--- VEÍCULOS CADASTRADOS ---")
    for i in range(len(veiculo["veiculos"])):
        print(f'{i+1} - {veiculo["veiculos"][i]} | Modelo: {veiculo["modelo"][i]} | Placa: {veiculo["placa"][i]}')
    print("-------------")

    placa_remover = input("Digite a placa do veículo que deseja remover: ").strip().lower()

    if placa_remover in veiculo["placa"]:
        indice = veiculo["placa"].index(placa_remover)
        veiculo["veiculos"].pop(indice)
        veiculo["modelo"].pop(indice)
        veiculo["placa"].pop(indice)
        print(f" Veículo com placa {placa_remover} removido com sucesso!")
    else:
        print(f" Placa {placa_remover} não encontrada!")

def listar(veiculo):
    if len(veiculo["veiculos"]) == 0:
        print(" Nenhum veículo cadastrado ainda!")
    else:
        print("--- LISTA DE VEÍCULOS ---")
        for i in range(len(veiculo["veiculos"])):
            print(f'{i+1} - {veiculo["veiculos"][i]} | Modelo: {veiculo["modelo"][i]} | Placa: {veiculo["placa"][i]}')
        print("-----------------------------")

def contar(veiculo):
    print(f" Total de veículos cadastrados: {len(veiculo['veiculos'])}")

def menu():
    while True:
        print("""
--- MENU DE CADASTRO DE VEÍCULOS ---
1 - Cadastrar veículo
2 - Remover veículo
3 - Listar veículos
4 - Ver quantidade cadastrada
5 - Sair
""")

        opcao = input("Escolha uma opção: ").strip()

        if opcao == "1":
            liscad(veiculo)
        elif opcao == "2":
            remover(veiculo)
        elif opcao == "3":
            listar(veiculo)
        elif opcao == "4":
            contar(veiculo)
        elif opcao == "5":
            print("Saindo do programa... ")
            break
        else:
            print("Opção inválida! Tente novamente.")

menu()
```
